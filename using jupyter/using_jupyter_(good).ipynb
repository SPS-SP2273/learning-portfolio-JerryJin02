# 3 Shell Magic

```python
# Python supports a series of 'shell magic'. 
# For instance, %timeit is a magical command that can tell you the runtime one line of code requires. 
# Similarly, %%timeit is another magical command that can tell the runtime required by a block of code. 
# When using %%timeit, one needs to be careful that this command must be keyed in as the first line of a cell \
#     in order for it too work. Otherwise, exception will arise.
```

```python
%timeit x=10
```

```python
%%timeit # First line of the cell
x=10
y=10
z=10
```
